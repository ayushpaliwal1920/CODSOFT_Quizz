```python
# fibonacci : using memoization : space ka khrcha krk time reduce krna

import time

def fib(n,d) :
   if n in d :
     return d[n]
   else :
    d[n] = fib(n-1,d) + fib(n-2,d)
    return d[n]
```

```python
start = time.time()
d = {0:1 ,1:1}  # dictionary
fib(100,d)
print(time.time() -start)
```
